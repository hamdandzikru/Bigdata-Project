```python
## Package
import datetime
import pymongo
import math
import pandas as pd
import json
import matplotlib.pyplot as plt
 
pd.set_option('display.max_columns', None)
```

```python
with open('category.json') as json_file:
    data = json.load(json_file)

category = pd.Series(data, index=['kind', 'etag', 'items'])

df_items = pd.DataFrame.from_records(category['items'])
df_id = df_items.iloc[:, [2]]

df_snippet = pd.DataFrame.from_records(df_items['snippet'])
df_title = df_snippet.iloc[:, [1]]

df_category = pd.concat([df_id, df_title], axis=1, sort=False)
df_category['category_id'] = df_category['category_id'].astype(int)
```

```python
## connect ke db
 
client = pymongo.MongoClient("mongodb://ilhamnyto:ilhamnyto@tugasbigdata-shard-00-00.hrhba.mongodb.net:27017,tugasbigdata-shard-00-01.hrhba.mongodb.net:27017,tugasbigdata-shard-00-02.hrhba.mongodb.net:27017/tugasbigdata?ssl=true&replicaSet=atlas-k403ha-shard-0&authSource=admin&retryWrites=true&w=majority")
db = client.tugasbigdata
cursor = db.youtube.find()
```

```python
data =  pd.DataFrame(list(cursor))
```

```python
data.drop(
    ['_id', 'video_id', 'channel_title', 'title', 'thumbnail_link', 'comments_disabled', 'ratings_disabled',
     'video_error_or_removed', 'descriptionn1WpP7iowLc'], axis=1, inplace=True)

data = data[data.trending_date != datetime.datetime(1970, 1, 1, 0, 0)]
data = data.dropna()

data['views'] = data['views'].astype(float).astype(int)
data['likes'] = data['likes'].astype(float).astype(int)
data['dislikes'] = data['dislikes'].astype(float).astype(int)
data['comment_count'] = data['comment_count'].astype(float).astype(int)
data['category_id'] = data['category_id'].astype(float).astype(int)
```

```python
df_data = pd.merge(data, df_category, on='category_id')

df_total = df_data.groupby(['trending_date', 'category'])[
    ['views', 'likes', 'dislikes', 'comment_count']].sum().reset_index()
type_date = df_data.trending_date.unique()
```

```python
df_total['views'] = df_total['views'].apply(lambda x: x * 1)
df_total['likes'] = df_total['likes'].apply(lambda x: x * 1.5)
df_total['dislikes'] = df_total['dislikes'].apply(lambda x: x * -1.5)
df_total['comment_count'] = df_total['comment_count'].apply(lambda x: x * 2)

df_total['strength'] = df_total.iloc[:, -4:].sum(axis=1)
```

```python
predict = df_total[df_total['trending_date'].isin(type_date[-7:])].reset_index()
predict = predict.groupby(['category'])[
    ['views', 'likes', 'dislikes', 'comment_count', 'strength']].sum().reset_index()
predict['strength_smooth'] = predict['strength'].apply(lambda x: math.log(1 + x, 2))
predict = predict.sort_values('strength', ascending=False)

```

```python
result = predict[['category', 'strength_smooth']].reset_index(drop=True)
print(result)
```

```python
# menampilkan prediksi 8 kategori yang trending 

plt.figure(figsize=(15,15))
ax1 = plt.subplot(211, aspect='equal')
predict.head(8).plot(kind='pie', y = 'strength', ax=ax1, autopct='%1.1f%%', 
 startangle=90,  labels=predict['category'], legend=False)

plt.show()
```
